## Data Cleaning

```python
train = train[['OriginalTweet', 'Sentiment']]
test = test[['OriginalTweet', 'Sentiment']]
```

```python
sentimentMap = {'Extremely Negative': -1, 'Negative': -1, 'Neutral': 0,
                'Positive': 1, 'Extremely Positive': 1}
train['sentimentScore'] = train['Sentiment'].map(lambda x: sentimentMap[x])
test['sentimentScore'] = test['Sentiment'].map(lambda x: sentimentMap[x])
```

```python
train.sentimentScore.value_counts().iloc[:10]
```

```python
# stemmer = PorterStemmer()


def cleanTweets(tweet):
    tweet = re.sub(r'\W+|\d+', ' ', tweet).lower()
    tweet = re.sub(r'http\S+', ' ', tweet)
    tweet = [word for word in tweet.split(
    ) if word not in stopwords.words('english')]
#    tweet = [stemmer.stem(word) for word in tweet] #Higher accuracy without stemming
    return tweet
```

```python
train['cleanTweet'] = train['OriginalTweet'].apply(lambda x: cleanTweets(x))
test['cleanTweet'] = test['OriginalTweet'].apply(lambda x: cleanTweets(x))
```

```python
train.head()
```

```python
X_train = train['cleanTweet']
```

```python
train.sentimentScore = pd.Categorical(train.sentimentScore)
train.sentimentScore = train.sentimentScore.cat.codes
# Turn response variable into one-hot response vector
y_train = to_categorical(train.sentimentScore)
```

```python
y_train
```

```python
test.head()
```

```python
X_test = test['cleanTweet']
```

```python
test.sentimentScore = pd.Categorical(test.sentimentScore)
test.sentimentScore = test.sentimentScore.cat.codes
# Turn response variable into one-hot response vector
y_test = to_categorical(test.sentimentScore)
```

```python
y_test
```

```python
#The maximum word features to be used
MAX_FEATURES = 10000
EMBEDDING_DIMS = 128
MAX_SEQUENCE_LEN = 47
```

```python
# Initialize and fit the tokenizer
tokenizer = Tokenizer(num_words=MAX_FEATURES,
                      filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True, split=' ')
tokenizer.fit_on_texts(X_train)
```

```python
num_words = len(tokenizer.word_index) + 1
```

```python
num_words
```

```python
X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test)
```

```python
X_train_padded_inputs = pad_sequences(X_train, MAX_SEQUENCE_LEN)
X_test_padded_inputs = pad_sequences(X_test, MAX_SEQUENCE_LEN)
```

## LSTM Multiclass Classification Modeling

```python
model = Sequential()
model.add(Embedding(input_dim=MAX_FEATURES, output_dim=EMBEDDING_DIMS,
          input_length=X_train_padded_inputs.shape[1]))
model.add(Dropout(0.2))
model.add(LSTM(256, activation='relu', dropout=0.2, return_sequences=True))
model.add(LSTM(256, activation='relu', dropout=0.2, return_sequences=True))
model.add(LSTM(256, activation='relu', dropout=0.2))
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer=Adam(),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss',
                               patience=3)

history = model.fit(X_train_padded_inputs, y_train,
                    batch_size=128, epochs=20, callbacks=[early_stopping],
                    validation_data=(X_test_padded_inputs, y_test))
```

## Model Evaluation

```python
plt.figure()
# Use the history metrics
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'])
plt.show()
```

```python
score = model.evaluate(X_test_padded_inputs, y_test, verbose=0)
print('Test Loss:', score[0])
print('Test Accuracy:', score[1])
```
